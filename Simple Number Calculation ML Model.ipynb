```python
#Import libraries
import keras
import numpy as np
```

##### Start the modeling process

```python
num_model = keras.Sequential([keras.layers.Dense(units=1, input_shape = [1])])
```

#### The compiling process and the Training process
##### Here we compile a few things.
##### This is based on a simple function where [data*3] + 4
##### Where we compile xs is the input, and ys is the output
##### The Mean Squared Error measures how close a regression line is to a set of data points. It is a risk function corresponding to the expected value of the squared error loss.


```python
num_model.compile(optimizer = 'sgd', loss='MSE')
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0],dtype=float)
ys = np.array([1.0, 4.0, 7.0, 10.0, 13.0, 16.0], dtype=float)
num_model.fit(xs, ys, epochs=300)
```

##### Here, we've trained the model for a total of 500 epochs, and with each epoch, the model gets more accurate and the error decreases
##### The Testing Phase

```python
data = 5.0
answer =num_model.predict([data])
print(answer)
```
